```python
import numpy as np
import matplotlib.pyplot as plt

def dispersion_relation(k, kappa1, kappa2, m, a):
    return np.sqrt((2 * kappa1 / m) * (1 - np.cos(k * a)) + (2 * kappa2 / m) * (1 - np.cos(2 * k * a)))

# Parámetros del sistema
kappa1 = 1.0  # Constante del resorte 1
kappa2 = 0.5  # Constante del resorte 2
m = 1.0      # Masa de los átomos
a = 1.0      # Separación entre átomos

# Rango de k en el que se graficará la relación de dispersión
k_values = np.linspace(-np.pi / a, np.pi / a, 1000)

# Calcula la relación de dispersión para cada valor de k
omega_values = dispersion_relation(k_values, kappa1, kappa2, m, a)

# Grafica la relación de dispersión
plt.plot(k_values, omega_values)
plt.title('Relación de Dispersión')
plt.xlabel('Número de onda (k)')
plt.ylabel('Frecuencia angular (omega)')
plt.grid(True)
plt.show()
```
