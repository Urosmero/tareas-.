```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #Importamos las librerías a usar.

df = pd.read_csv('river-level.csv', index_col=0, comment='#', parse_dates=True) #Leemos el csv de los niveles del río agregando el comando "parse_dates" para reconocer las fechas

muestramensual = df.resample('M') #Creamos una muestra que tome los meses de nuestro pandas

df['avg_level'].plot(label='Promedio diario') # Y finalmente hacemos una gráfica para el promedio diario
muestramensual['avg_level'].mean().plot(label='Promedio Mensual') #para el promedio mensual
muestramensual['min_level'].min().plot(label='Mínimo mensual')#mínimos
muestramensual['max_level'].max().plot(label='Máximo mensual')#máximos
 
plt.xlabel('Fecha')
plt.ylabel('Nivel del río')
plt.legend()
plt.show()#Y graficamos
```
